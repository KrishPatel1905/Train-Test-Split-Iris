```python
import pandas as pd


from sklearn.ensemble  import RandomForestClassifier 
```

```python
data = pd.read_excel("iris-test.xlsx")

data.head()
```

```python
x = data.iloc[:,:-1] #fillter Row and Coloum

y = data.iloc[:,-1] #reaming row and coloums
```

```python
model = RandomForestClassifier()


model.fit(x.values ,y.values)


predictions = model.predict([[5.4,	3,	4.5,	1.5]])

print(predictions)
```

```python
datatest = pd.read_excel("iris-test.xlsx")

x_test = datatest.iloc[:, :-1]
predictions = model.predict(x_test.values)
print(predictions)
```
